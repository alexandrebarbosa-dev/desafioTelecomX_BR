#📌 Extracão

Para iniciar sua análise, você precisará importar os dados da API da Telecom X. Esses dados estão disponíveis no formato JSON e contêm informações essenciais sobre os clientes, incluindo dados demográficos, tipo de serviço contratado e status de evasão.

O que você deve fazer:

✅ Carregar os dados diretamente da API utilizando Python.

```python
import pandas as pd
```

```python
url = 'https://raw.githubusercontent.com/ingridcristh/challenge2-data-science/refs/heads/main/TelecomX_Data.json'
dados = pd.read_json(url)
dados.head()
```

## Tratando as Inconsistências

```python
(df['Churn'] == '').sum()
```

```python
df['Churn'] = df['Churn'].replace('', pd.NA)
```

```python
(df['Churn'] == '').sum()
```

```python
df[colunas_numericas].describe()
```

```python
df['Churn'].unique()
```

```python
df = df.dropna(subset=['Churn'])
```

```python
df['Churn'].unique()
```

```python
df['account.Charges.Total'].unique()
```

```python
df['account.Charges.Total'] = pd.to_numeric(
    df['account.Charges.Total'],
    errors='coerce'
)
```

```python
df['account.Charges.Total']
```

Ao tentar converter a coluna Charges.Total para numérica, retornou um erro, indicando strings vazias. Ao fazer a análise do DF como um todo, verifiquei que isso aconteceu porque a coluna customer.tenure é igual a zero, ou seja, este é um cliente que acabou de fazer sua assinatura, não tendo gerado valores ainda. Logo, o ideal para a coluna Charges.Total é que seja igual a zero também, em vez de remover, por exemplo.

```python
df['account.Charges.Total'] = df['account.Charges.Total'].astype(float)
```

## Padronização e Transformação de Dados (Opcional)

```python
colunas_relevantes = [
    'Churn',
    'customer.gender',
    'customer.SeniorCitizen',
    'customer.Partner',
    'customer.Dependents',
    'customer.tenure',
    'account.Contract',
    'account.PaymentMethod',
    'internet.TechSupport',
    'account.Charges.Monthly',
    'account.Charges.Total'
]
```

```python
df_analise = df[colunas_relevantes].copy()
df_analise.head()
```

```python
colunas_binarias = []

for col in df.columns:
    if df[col].dtype == 'object':
        valores = set(df[col].dropna().unique())
        if valores == {'Yes', 'No'}:
            colunas_binarias.append(col)

colunas_binarias
```

```python
for col in colunas_binarias:
    print(col, df[col].isna().sum())
```

```python
for col in colunas_binarias:
    df[col] = df[col].map({'Yes': 1, 'No': 0}).astype('int64')
```
