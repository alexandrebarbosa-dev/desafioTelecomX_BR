```python
colunas_binarias = []

for col in df.columns:
    if df[col].dtype == 'object':
        valores = set(df[col].dropna().unique())
        if valores == {'Yes', 'No'}:
            colunas_binarias.append(col)

colunas_binarias
```

```python
for col in colunas_binarias:
    print(col, df[col].isna().sum())
```

```python
for col in colunas_binarias:
    df[col] = df[col].replace({'Yes': 1, 'No': 0}).astype('int8')
```

```python
faixas = df_analise.copy()
faixas
```

```python
bins = [0, 12, 24, 36, 60, 72]

faixas = pd.cut(
    df_analise['Meses de contrato'],
    bins=bins
)

df_analise.head()
```

```python
tabela_frequencias = faixas.value_counts().sort_index().reset_index()
tabela_frequencias.columns = ['faixa de tempo de contrato', 'frequencia']

tabela_frequencias['porcentagem'] = (
    tabela_frequencias['frequencia'] / len(df_analise)
) * 100

tabela_frequencias
```

```python
tabela_churn = pd.crosstab(faixas, df_analise['Churn'])
tabela_churn
```

```python
tabela_percentual = pd.crosstab(
    faixas,
    df_analise['Churn'],
    normalize='index'
) * 100
tabela_percentual.columns = ['Ficou', 'Saiu']
tabela_percentual
```

```python
fig, ax = plt.subplots(figsize=(8,5))

tabela_percentual.plot(
    kind='bar',
    ax=ax,
    color=['royalblue', 'indianred'],
    edgecolor='black'
)

ax.set_title('Taxa de Evasão por Faixa de Tempo de Contrato')
ax.set_xlabel('')
ax.set_ylabel('Percentual (%)')
ax.legend(title='Situação')

plt.xticks(rotation=0)

plt.tight_layout()
plt.show()
```

```python
#Pedi à IA para aperfeiçoar o gráfico.

fig, ax = plt.subplots(figsize=(10,6))

# Criando labels mais claras
labels = [f'de {int(interval.left)} a {int(interval.right)} meses'
          for interval in tabela_percentual.index]

x = np.arange(len(labels))
largura = 0.35

# Barras
barras1 = ax.bar(x - largura/2, tabela_percentual['Ficou'],
                 largura, label='Ficou',
                 color='royalblue', edgecolor='black')

barras2 = ax.bar(x + largura/2, tabela_percentual['Saiu'],
                 largura, label='Saiu',
                 color='indianred', edgecolor='black')

# Inserir valores percentuais
for barras in [barras1, barras2]:
    for barra in barras:
        altura = barra.get_height()
        ax.text(barra.get_x() + barra.get_width()/2,
                altura,
                f'{altura:.1f}%',
                ha='center', va='bottom', fontsize=9)

# Ajustes visuais
ax.set_title('Taxa de Evasão por Faixa de Tempo de Contrato')
ax.set_xlabel('Tempo de contrato')
ax.set_ylabel('Percentual (%)')
ax.set_xticks(x)
ax.set_xticklabels(labels, rotation=0)
ax.legend()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.set_facecolor('lightgray')


plt.tight_layout()
plt.show()
```
