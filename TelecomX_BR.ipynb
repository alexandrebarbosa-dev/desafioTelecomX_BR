## Tratando as Inconsistências

```python
(df['Churn'] == '').sum()
```

```python
df['account.Charges.Total'] = pd.to_numeric(
    df['account.Charges.Total'],
    errors='coerce'
)
```

```python
df['account.Charges.Total']
```

Ao tentar converter a coluna Charges.Total para numérica, retornou um erro, indicando strings vazias. Ao fazer a análise do DF como um todo, verifiquei que isso aconteceu porque a coluna customer.tenure é igual a zero, ou seja, este é um cliente que acabou de fazer sua assinatura, não tendo gerado valores ainda. Logo, o ideal para a coluna Charges.Total é que seja igual a zero também, em vez de remover, por exemplo.

```python
df['account.Charges.Total'] = df['account.Charges.Total'].astype(float)
```

## Padronização e Transformação de Dados (Opcional)

```python
colunas_relevantes = [
    'Churn',
    'customer.gender',
    'customer.SeniorCitizen',
    'customer.Partner',
    'customer.Dependents',
    'customer.tenure',
    'account.Contract',
    'account.PaymentMethod',
    'internet.TechSupport',
    'account.Charges.Monthly',
    'account.Charges.Total'
]
```

```python
df_analise = df[colunas_relevantes].copy()
df_analise.head()
```

```python
colunas_binarias = []

for col in df.columns:
    if df[col].dtype == 'object':
        valores = set(df[col].dropna().unique())
        if valores == {'Yes', 'No'}:
            colunas_binarias.append(col)

colunas_binarias
```

```python
for col in colunas_binarias:
    print(col, df[col].isna().sum())
```

```python
for col in colunas_binarias:
    df[col] = df[col].replace({'Yes': 1, 'No': 0}).astype('int8')
```

```python
df_analise.head()
```

```python
df_analise = df_analise.rename(columns={
    'customer.gender': 'Genero',
    'customer.SeniorCitizen': 'Idoso',
    'customer.Partner': 'Parceiro',
    'customer.Dependents': 'Dependentes',
    'customer.tenure': 'Meses de contrato',
    'account.Contract': 'Tipo de contrato',
    'account.PaymentMethod': 'Forma de pagamento',
    'internet.TechSupport': 'Suporte tecnico',
    'account.Charges.Monthly': 'Total mensal',
    'account.Charges.Total': 'Total geral'
})
```

# Distribuição da Evasão

```python
churn_contagem = df_analise['Churn'].value_counts(normalize=True)
```

```python
import matplotlib.pyplot as plt
```

```python
import matplotlib.pyplot as plt

labels = ['Permaneceu', 'Saiu']

plt.figure(figsize=(6,6))
plt.pie(
    churn_contagem,
    labels=labels,
    autopct='%1.1f%%',
    startangle=90
)

plt.title('Distribuição de Evasão de Clientes')
plt.show()
```

# Contagem de Evasão por Variáveis Categóricas

Montagem dos gráficos. Opção 1

```python
churn_genero = df_analise.groupby('Genero')['Churn'].mean() * 100
```

```python
churn_tipo_contrato = df_analise.groupby('Tipo de contrato')['Churn'].mean() * 100
```

```python
churn_suporte = df_analise.groupby('Suporte tecnico')['Churn'].mean() * 100
```

```python
churn_forma_pagamento = df_analise.groupby('Forma de pagamento')['Churn'].mean() * 100
```

```python
fig, ax = plt.subplots(ncols=2, nrows=2, figsize=(12, 8))
churn_genero.plot(kind='bar', ax=ax[0,0], color='gray')
ax[0,0].set_title('Evasão por Gênero')
ax[0,0].set_xlabel('Gênero')
ax[0,0].set_ylabel('Taxa de Evasão')
for i, v in enumerate(churn_genero):
    ax[0,0].text(i, v + 0.01, f'{v:.2f}%', ha='center')
churn_forma_pagamento.plot(kind='bar', ax=ax[0,1], color='gray')
ax[0,1].set_title('Evasão por Forma de pagamento')
ax[0,1].set_xlabel('Forma de pagamento')
ax[0,1].set_ylabel('Taxa de Evasão')
for i, v in enumerate(churn_forma_pagamento):
    ax[0,1].text(i, v + 0.01, f'{v:.2f}%', ha='center')
churn_suporte.plot(kind='bar', ax=ax[1,0], color='gray')
ax[1,0].set_title('Evasão entre quem tem e quem não tem suporte')
ax[1,0].set_xlabel('Tem suporte?')
ax[1,0].set_ylabel('Taxa de Evasão')
for i, v in enumerate(churn_suporte):
    ax[1,0].text(i, v + 0.01, f'{v:.2f}%', ha='center')
churn_tipo_contrato.plot(kind='bar', ax=ax[1,1], color='gray')
ax[1,1].set_title('Evasão por Tipo do contrato')
ax[1,1].set_xlabel('Tipo do contrato')
ax[1,1].set_ylabel('Taxa de Evasão')
for i, v in enumerate(churn_tipo_contrato):
    ax[1,1].text(i, v + 0.01, f'{v:.2f}%', ha='center')
plt.tight_layout()
plt.show()
```

Opção 2 - cada gráfico em uma figura

```python
fig, ax = plt.subplots(figsize=(8,6))
churn_genero.plot(kind='bar', color='gray', ax=ax)
ax.set_title('Evasão por Gênero')
ax.set_xlabel('Gênero')
ax.set_ylabel('Taxa de Evasão (%)')

# Colocar os valores em cima das barras
for i, v in enumerate(churn_genero):
    ax.text(i, v + 0.3, f'{v:.2f}%', ha='center')

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8,6))
churn_suporte.plot(kind='bar', color='gray', ax=ax)
ax.set_title('Evasão entre quem tem e quem não tem suporte')
ax.set_xlabel('Tem suporte?')
ax.set_ylabel('Taxa de Evasão (%)')

# Colocar os valores em cima das barras
for i, v in enumerate(churn_suporte):
    ax.text(i, v + 0.3, f'{v:.2f}%', ha='center')

plt.show()
```
